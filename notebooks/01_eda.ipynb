# 01 — Exploratory analysis

Load the canonical parquet produced by `scripts/ingest.py` and inspect hardness vs. volume fraction, routes, and missingness.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

root = Path("..").resolve()
df = pd.read_parquet(root / "data/interim/canonical.parquet")
df.head()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.scatterplot(data=df, x="vol_frac_reinf", y="hv", hue="matrix_formula", ax=ax)
ax.set_title("Vickers HV vs. reinforcement volume fraction")
plt.tight_layout()
```
